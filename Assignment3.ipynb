```python
data={
    "Age": [16,25,30,35,40,45,50],
   "Income":[2333,334,3455,344,667,7666,5677],
    "Age Group":['10-23','23-45','56-66' ,'45-56' ,'90-34' ,'23-23' ,'12-34']}
```

```python
data
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
summary_stats=df.groupby('Age Group')['Income'].describe()
```

```python
summary_stats
```

```python
summary_stats=df.groupby('Age Group')['Income'].std()
```

```python
summary_stats
```

```python
summary_stats=df.groupby('Age Group')['Income'].mean()
```

```python
summary_stats
```
