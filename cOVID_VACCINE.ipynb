```python
#Number of persons state-wise vaccinated for the First Dose

# Remove rows like 'India' total or empty states if any
df_States=df[df['State']!='India']
```

```python
df_first_dose=df_States.groupby('State')['First Dose Administered'].sum().sort_values(ascending = False)
```

```python
df_first_dose
```

```python
#Number of persons state-wise vaccinated for the Second Dose
second_dose=df_States.groupby('State')['Second Dose Administered'].sum().sort_values(ascending=False)
```

```python
second_dose
```

```python
#Number of Males vaccinated
Male_vaccinated=df_States.groupby('State')['Male(Individuals Vaccinated)'].sum().sort_values(ascending=False)
```

```python
Male_vaccinated
```

```python
female_vaccinated = df_States.groupby('State')['Female(Individuals Vaccinated)'].sum().sort_values(ascending=False)

print("Female Vaccinated (state-wise):")
print(female_vaccinated)
```

```python
df_first_dose.head(10).plot(kind='bar', figsize=(10,5), title='Top 10 States - First Dose Vaccination')
plt.ylabel('Number of People')
plt.xlabel('States')
plt.show()
```

```python
second_dose.head(10).plot(kind='bar',figsize=(10,5),title="Seccond dose distribution")
plt.ylabel("no . of people")
plt.xlabel("states")
plt.show()
```
