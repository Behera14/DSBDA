```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = sns.load_dataset('iris')  # or use pd.read_csv() if using a local file

# Plot histograms for all numeric features
df.hist(figsize=(10, 8), bins=20, color='skyblue', edgecolor='black')

plt.suptitle('Histograms of Iris Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python

 # Drop the categorical column for boxplot
 plt.figure(figsize=(10, 6))
 sns.boxplot(data=df.drop(columns=['species']), palette='Set3')
 plt.title('Boxplots of Iris Features', fontsize=16)
 plt.grid(True)
 plt.show()
```

```python
Q1 = df['sepal_width'].quantile(0.25)
Q3 = df['sepal_width'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['sepal_width'] < Q1 - 1.5 * IQR) | (df['sepal_width'] > Q3 + 1.5 * IQR)]
print(outliers)
```
